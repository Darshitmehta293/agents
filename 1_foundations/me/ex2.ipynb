```python
from openai import OpenAI
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="anything")
```

```python
content = "Please come up with a challenging, nuanced question that I can ask a number of LLMs to evaluate their intelligence. "
content += "Answer only with the question, no explanation."
messages = [{"role": "user", "content": content}]
```

```python
response = ollama.chat.completions.create(
    model = "llama3.2",
    messages = messages
)
content = response.choices[0].message.content
print(content)
```

```python
competitors=[]
answers=[]
messages = [{"role": "user", "content": content}]
```

```python
model_name = "llama3.2"
response = ollama.chat.completions.create(model=model_name,messages=messages)
answerz = response.choices[0].message.content
display(Markdown(answerz))
competitors.append(model_name)
answers.append(answerz)
```

```python
print(competitors)
print(answers)
```

```python
for competitors, answer in zip(competitors, answers):
    print(f"Competitor: {competitor}\n\n{answer}")
```
